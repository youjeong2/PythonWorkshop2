```python
# Python 수치해석 라이브러리
import numpy as np

# 회귀 분석용 테스트 데이터 만들어주는 라이브러리
from sklearn.datasets import make_regression

#데이터 구성하기
X, y = make_regression(
    n_samples = 500, n_features = 2,
    n_informative = 2, noise = 5, 
    random_state = 0 # random_state는 자율로 정하기
)

print(X.shape) #500개 2차원
print(y.shape)
```

```python
#dummy의 범위 정하는 것
print("output range = [%f %f]" %(min(y), max(y)))
```

```python
# pip install keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
def model_build():
    model = Sequential()
    # 컨벌루션 뉴럴네트워크 설정 3개
    model.add(
        Dense(16, input_dim = 2, activation = "relu")
    )
    model.add(Dense(8, activation = "relu"))
    model.add(Dense(1))
    
    model.compile(
        loss='mean_squared_error', optimizer='adam' # loss는 최소제곱오차
    )
    
    return model
```

```python
# Regressor 써보기 
# 회귀분석 모델 생성해줌
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
YourModel = KerasRegressor(
    build_fn = model_build,
    epochs = 400,
    batch_size = 10,
    verbose = 1
)
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5)
```

```python
#모델이 간단하고, 하이퍼파라미터가 많이 없으면 빠르게 끝남
from sklearn.model_selection import cross_val_score

results = cross_val_score(YourModel, X, y, cv=kf)
```

```python
print("Final Cross Validation Loss:", abs(results.mean()))
# 교차검증 한 것에 대한 전체 loss 값의 평균 
# loss가 25로 값이 크게 나옴
```

```python
# 5장 결론
# l2
# l2는 오차에 대한 에러율을 ㅏㅈ추는 데 활용한다.
# 오차의 제곱을 가지고 연산하므로 잘못된 정보를 판단 할 가능성이 낮아진다.
# 특히 음수와 양수 형태로 구성된 오차간에 저보 오판율은 굉장히 낮다.
# l2 단점 : 제곱을 가지고 하는 녀석들의 특성 -> 큰값으로 제곱하면 한없이 커지고, 작은값으로 하면 한없이 작아짐
# 분산값이 굉장히 커질 수 있다.
# 그래서 l1과 함께 쓰는 것
# 분산이 커지는 것이 싫다면 l1, l2같이 쓰기

# Dropout은 잡음에 대한 정보 학습하는 과적합을 해결하기 위한 녀석
# 그러므로 일정 수치 이하의 정보에 대해서는 걸러내도록 만든다.
# 이 수치를 너무 크게 줄 경우 실제 정보도 같이 증발되므로
# 적정한 수치 값을 찾아내는 것이 매우 중요
```

```python
# train은 훈련할 수록 loss가 작아짐
# test는 훈련(=학습)할 수록 loss가 커짐
# -> 과소적합(학습데이터 늘리기), 과적합(정규화(오차를최소화), dropout(영향미치지 않는 잡음제거)
# 훈련을 더 많이 했더니 loss가 커짐 -> 훈련하면서 잡음까지 학습되었다.
# 진짜 훈련은 초반에 끝남 그 후는 잡음학습
```

```python
# 훈련을 많이 했더니 훈련 로스는 없다.
# 그런데 테스트 로스가 늘어난다?
# 결국 과적합이 되어서 노이즈 정보까지 처리를 하는 상황

# 해결책
# 1. 훈련 수를 줄이면 된다[EarlyStopping}- 조기중단
# 2. l2를 사용해서 오차의 제곱의 합을 가지고 연산
#    오차가 양수(+), 음수(-)로 인해 잘못된 판정을 할 수 있다.
# 그러므로 양쪽 모두 양수가 될 수 있는 제곱형태를 취함
# Dropout은 일정 확률 이하의 정보를 폐기 처분
```

```python
# 조기 중단을 할 수 있게 임포트
from keras.callbacks import EarlyStopping
```

```python
# 조기중단 예제 만들어보기 

model = Sequential()

# 네트워크(=뉴럴넷) 구성
model.add(Dense(16,activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'sgd'
)

# EarlyStopping을 사용한다면
# Fit을 하기 이전에 어디서 중단할지를 정해줘야 한다.
es_callback  = EarlyStopping(
    #val_loss : 테스트 로스를 감독함
    monitor = 'val_loss',
    # 최소값 감지
    mode = 'min'
) # 테스트 로스 보다가 높아지면 그 전의 최소값을 받을 수 있게 callback 넣은 것

# 콜백이란? 이벤트
# 몰컴 ㅋ ㅋ ㅋ 
# 1.부모님이 주무시고 계셔야 한다. 
# 2. 큰 소리가 나면 안됨
# 3. 부모님이 방문을 열고 나오실 때
#  모니터 끄고 자는 척 시전 -> callback 
# 예상치 못한 사건이 발생해야함 
 
# 피팅할 때 조건이 만족되면 콜백에 의해 fit이 중단되도록 구성한 부분
trainRes = model.fit(
    X_train, y_train, 
    validation_data = (X_test, y_test),
    epochs = 400,
    batch_size = 5,
    # 조건이 만족되면 콜백에 의해 fit이 중단되도록 구성한 부분
    callbacks = [es_callback],
    verbose = 0
)

print("Test Loss =", model.evaluate(X_test, y_test))

# min값을 찍으면 끝내라 
```

```python
plt.plot(trainRes.history['loss'])
plt.plot(trainRes.history['val_loss'])
plt.ylim(0, 1)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')
```

```python
model = Sequential()

model.add(Dense(16, input_dim = 16, activation = 'relu'))
model.add(GaussianNoise(0.01))
model.add(Dropout(0.5))
model.add(Dense(8, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(4, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'sgd'
)

trainRes = model.fit(
    X_train, y_train,
    batch_size = 5,
    epochs = 400, 
    verbose = 0
)
```
