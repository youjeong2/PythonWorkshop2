```python
#  지금 까지 한것들은 정보를 yes, no로 보고 있음
# 한번에 볼 수 있는 방법
# p41

# 위에서 힘들게 람다를 쓰고 아주 난리를 쳤는데
# pd.get_dummies 한 방으로 모든것이 끝났다.
marital_dummies = pd.get_dummies(bank_data['marital'])

marital_dummies.tail(20)
```

```python
# 4507 번째 사람이 이혼을 했다는 것을 알 수 있음
# pd.get_dummies 라는 라이브러리를 쓰면 한방으로 끝남 lamda 쓰는 것 보다 훨씬 편함


# p57 케라스단점 라이브러리 그대로 써서 깊이 있게 학습하기어려움, 커스텀이 어려움 
# 더 깊숙히 쓰기 위해서는 수학을 공부해야함
```

```python
# get_dummies()로 얻은 결과도 여전히 drop()을 통해서
# 정보 자체를 배제할 수 있다.
marital_dummies.drop('divorced', axis=1, inplace=True)
marital_dummies.columns = [
    f'marital_{colname}' 
    for colname in marital_dummies.columns
]
marital_dummies.tail(20)
```
